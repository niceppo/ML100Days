```python
import os
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv('application_train d008.csv')
train_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
train_data.describe()
```

```python
d_df=train_data.dtypes.reset_index()
d_df.columns=['Count','Type']
d_df=d_df.groupby('Type').agg('count').reset_index()
d_df
```

```python
int_b=[]
float_b=[]
obj_b=[]
other_b=[]

for dtype,feature in zip(train_data.dtypes,train_data.columns):
    if dtype=='int64':
        int_b.append(feature)
    elif dtype=='float64':
        float_b.append(feature)
    elif dtype=='object':
        obj_b.append(feature)
    else: other_b.append(feature)

        
print(f'{len(int_b)} Integer Features: {int_b}\n')
print(f'{len(float_b)} Float Features: {float_b}\n')
print(f'{len(obj_b)} Object Features: {int_b}\n')
print(f'{len(other_b)} Other Features: {other_b}')
```

```python
train_data[obj_b].describe()
```

```python
print('Max=',train_data['CNT_FAM_MEMBERS'].max())
print('Min=',train_data['CNT_FAM_MEMBERS'].min())
print('Mean=',train_data['CNT_FAM_MEMBERS'].mean())
print('Std=',train_data['CNT_FAM_MEMBERS'].std())
```
