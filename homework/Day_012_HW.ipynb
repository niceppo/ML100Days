```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
train_data=pd.read_csv('Lost_r_standardize_HW_012/titanic_train.csv')
test_data=pd.read_csv('Lost_r_standardize_HW_012/titanic_test.csv')
print(train_data.head())
print(test_data.head())
```

```python
#save 'survived' to train_y before drop it
train_Y=train_data['Survived']
train_data=train_data.drop(['PassengerId','Survived'],axis=1)
test_data=test_data.drop(['PassengerId'], axis=1)

df=pd.concat([train_data,test_data])
df.head()
```

```python
df.shape
```

```python
from collections import defaultdict
```

```python
# Ticket categories
mode_dict=defaultdict(lambda:0)

for value in df[~df['Ticket'].isnull()]['Ticket']:
    mode_dict[value] +=1
mode_get=sorted(mode_dict.items(), key= lambda kv :kv[1], reverse=True)
print(mode_get)
```

```python
cabin_dict=defaultdict(lambda:0)

for value in df[~df['Cabin'].isnull()]['Cabin']:
    cabin_dict[value] +=1
cabin_get=sorted(mode_dict.items(), key= lambda kv :kv[1], reverse=True)
print(cabin_get)
```

```python
cabin_df=pd.DataFrame.from_dict(cabin_get)
cabin_df.columns=['Cabin_name','Cabin_num']
cabin_df.head()
```

```python
cabin_df=df.groupby('Cabin').agg('count').reset_index()
cabin_df
```

```python
cabin_df['Max']=cabin_df[['Pclass','Name']].max(axis=1)
cabin_df.head()
```

```python
cabin_df=cabin_df.drop(['Pclass','Name'], axis=1)
```

```python
df['Cabin_num'] = df['Cabin'].map(cabin_df.set_index('Cabin')['Max'])
df.head()
```

```python
embarked_df=df.groupby('Embarked').agg('count')
embarked_df.head()
```

```python
c=0
q=0
s=0
for i in df['Embarked']:
    if i=='C':
        c +=1
    elif i=='Q':
        q+=1
    elif i=='S':
        s +=1
print(c,q,s)
```

```python
embarked_dict=[('C',270,1),('Q',123,2),('S',914,3)]
embarked_df=pd.DataFrame.from_dict(embarked_dict)
embarked_df.columns=['Embarked','Count','EM_ID']
embarked_df
```

```python
df['Embarked_ID']=df['Embarked'].map(embarked_df.set_index('Embarked')['EM_ID'])
df.head()
```

```python
num_feature=[]

for dtype, feature in zip (df.dtypes, df.columns):
    if dtype=='float64' or dtype=='int64':
        num_feature.append(feature)

print(f'{len(num_feature)} Numeric Features:{num_feature}')
```

```python
df_n=df[num_feature]
df_n.head()
```

```python
train_Y.shape[0]
```

```python
#na fill in -1
df_n_m1=df_n.fillna(-1)
train_num=train_Y.shape[0]
train_X=df_n_m1[:train_num]
estimator=LogisticRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
#na fill in 0
df_n_0=df_n.fillna(0)

train_X=df_n_0[:train_num]

cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df_o=df_n.drop(['Cabin_num','Embarked_ID'],axis=1)
df_o
```

```python
#na fill in -1
df_n_m1=df_o.fillna(-1)
train_num=train_Y.shape[0]
train_X=df_n_m1[:train_num]
estimator=LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df_n_0T=MinMaxScaler().fit_transform(df_n_0)
train_X=df_n_0T[:train_num]
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df_n_0s=StandardScaler().fit_transform(df_n_0)
train_X=df_n_0s[:train_num]
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
